## 1. Theoretical Knowledge Preparation



### 1.1 Understanding Types of Multimodal Large Language Models

- Classification of existing multimodal large language models by functionality and modality support


> Before constructing MLLMs，researchers have reached consensus、a key premise：Due to scaling laws and emergent phenomena，Current language-based LLMs possess strong semantic understanding，which meanslanguage is a key modality for carrying intelligence，language intelligence is the hub of multimodal intelligence。Therefore，almost all MLLMs build on language-based LLMs，with this LLM as the core decision-making module，similar to a brain or central processor。In other words，by adding additional external non-text modal modules or encoders，LLM is given multimodal perception/operation capabilities。

We classify existing MLLMs by modality and functional support。

### 1.2 Understanding the General Technical Framework


- Architecture 1: LLM as Task Scheduler

> Currently two common MLLM architectures exist。
First type is“LLMas discrete scheduler/controller”architecture。As shown below，LLM receives text signals and sends commands to downstream modules。All system message passing uses LLM text commands。Functional modules do not interact。


![Architecture1](assets/Architecture1.png)

- Architecture 2: LLM as Joint Part of System

> Second: encoder-LLM-decoder framework。
This is the most popular architecture today。LLM perceives multimodal info and responds within encoder-LLM-decoder structure。Therefore，Key difference from first architecture:：LLM as key joint system part receives multimodal info and delegates to decoder。In encoder-LLM-decoder framework，As shown below，encoder processes signals from multiple modalities，LLMserves as core decision maker，decoder manages multimodal outputs。

![Architecture2](assets/Architecture2.png)

#### 2.5.2 Embedding Preparation

In NExT-GPT decoder alignment, we minimize Signal Token and caption distances。To ensure efficiency and save time/memory, we precompute embeddings using diffusion model encoders。

Before training NExT-GPT, run this command to save embedding files。
```
cd ./code/
python process_embeddings.py ../data/T-X_pair_data/cc3m/cc3m.json image ../data/embed/ runwayml/stable-diffusion-v1-5
```


Parameter description:：
- args[1]: caption file path;
- args[2]: modality: image, video or audio;
- args[3]: embedding file save path;
- args[4]: pretrained diffusion model name。




#### 2.5.3 Three-Stage Training


Refer to base.yaml to understand module system settings。

Then run this script to start NExT-GPT training:：
```
cd ./code
bash scripts/train.sh
```

Specify command as follows:：
```
deepspeed --include localhost:0 --master_addr 127.0.0.1 --master_port 28459 train.py \
    --model nextgpt \
    --stage 1\
    --save_path  ../ckpt/delta_ckpt/nextgpt/7b_tiva_v0/\
    --log_path ../ckpt/delta_ckpt/nextgpt/7b_tiva_v0/log/
```
Key parameters:：
- `--include`: localhost:0 means GPU 0 in DeepSpeed。
- `--stage`: training stage。
- `--save_path`: directory for delta weights (auto-created)。
- `--log_path`: log file directory。



NExT-GPT training has 3 steps:：

- **Step 1**: Encoder-side LLM-centered multimodal alignment. Train input projection layers, freeze ImageBind/LLM/output layers.
  
  Run train.sh with --stage 1
  
  Refer to stage_1.yaml and deepspeed configs for more info。

Note: dataset names must match catalog.py definitions。



- **Step 2**: Decoder-side instruction-following alignment. Train output layers, freeze ImageBind/LLM/input layers.

  Simply run the above`train.sh`script，and set：`--stage 2`

Refer to stage_2.yaml and deepspeed configs for more info。



- **Step 3**: Instruction tuning. Fine-tune LLM with LoRA and adjust projection layers.

  Simply run the above`train.sh`script，and set：`--stage 3`

Refer to stage_3.yaml and deepspeed configs for more info。
